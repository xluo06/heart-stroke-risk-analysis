```python
HD_stroke_list = []
for i in range(len(df)):
    if(df['HeartDiseaseorAttack'][i]==1 and df['Stroke'][i]==1):
        HD_stroke_list.append('(+,+)')
    elif(df['HeartDiseaseorAttack'][i]==1 and df['Stroke'][i]==0):
        HD_stroke_list.append('(+,-)')
    elif(df['HeartDiseaseorAttack'][i]==0 and df['Stroke'][i]==1):
        HD_stroke_list.append('(-,+)')
    else:
        HD_stroke_list.append('(-,-)')
```

```python
df1 = df.copy()
```

```python
df1['HD_stroke'] = HD_stroke_list
```

```python
df1_crosstab = pd.crosstab(df1['HD_stroke'],df1['HighBP'],margins=True, margins_name='Total')
df1_crosstab
```

```python
# drop 'HeartDiseaseorAttack' and 'Stroke' column replaced 'HD_stroke' as target variable 
columns_to_drop = ['HeartDiseaseorAttack','Stroke']
df_drop_HD = df1.drop(columns_to_drop, axis=1)
```

```python
# Calculate the conditional entropy of each of the other columns given the response variable
conditional_entropy = {}
for col in df_drop_HD.columns:
    if col != 'HD_stroke':
        conditional_entropy[col] = mutual_info_score(df_drop_HD['HD_stroke'], df_drop_HD[col])
```

```python
conditional_entropy
```

```python
# Sort the conditional entropy values from lowest to highest
sorted_conditional_entropy = sorted(conditional_entropy.items(), key=lambda x: x[1])
# Print the entropy and conditional entropy values in sorted order
for col, entropy in sorted_conditional_entropy:
    print(col, entropy)
```
