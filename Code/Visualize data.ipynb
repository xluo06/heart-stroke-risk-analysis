```python
import matplotlib.pyplot as plt

# Example data
x = ['0,1', '0,10', '0,11', '0,12', '0,2', '0,20', '0,3', '0,4']
y = [1, 10, 11, 12, 2, 20, 3, 4]

# Define custom sorting function
def custom_sort(label):
    parts = label.split(',')
    return int(parts[0]), int(parts[1])

# Sort x-axis values and corresponding y-axis values using custom sorting function
x_sorted, y_sorted = zip(*sorted(zip(x, y), key=lambda pair: custom_sort(pair[0])))

# Create plot with sorted x-axis
plt.plot(x_sorted, y_sorted)

# Add x-axis label
plt.xlabel('x')

# Display plot
plt.show()
```

```python
df_copy = df.copy()

# define the bin edges and labels
bins = [0, 18.5, 24.9, 29.9, float("inf")]
labels = [1, 2, 3, 4]

# categorize the BMI column into 4 levels
df_copy["BMI"] = pd.cut(df_copy["BMI"], bins=bins, labels=labels)
df_copy.head()
```

```python
# Loop through each column in n_df and create heatmap
for col in df_copy.columns:
    # Create contingency table
    contingency_table = pd.crosstab(df['HeartDiseaseorAttack'], df_copy[col], normalize='columns')
    plt.figure(figsize = (8,6))
    # Create heatmap using seaborn
    sns.heatmap(contingency_table, cmap='viridis_r')
    plt.title("HD vs " + col)
    plt.show()
```
